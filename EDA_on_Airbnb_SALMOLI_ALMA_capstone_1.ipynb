# EDA

```python
sns.boxplot(x='neighbourhood_group', y='price',data=df)
```

```python
sns.boxplot(x='neighbourhood', y='price',data=df)
```

### Neighbourhood group

```python
plt.style.use('fivethirtyeight')
plt.figure(figsize=(13,7))
plt.title("Neighbourhood Group")
g = plt.pie(df.neighbourhood_group.value_counts(), labels=df.neighbourhood_group.value_counts().index,autopct='%1.1f%%', startangle=180)
plt.show()
```

### *INSIGHT 1--> The pie chart above shows that Airbnb Listings in Newyork are near Manhattan, and Brooklyn has the highest share of hotels. 

## Room type

```python
plt.figure(figsize=(13,7))
plt.title("Type of Room")
sns.countplot(df.room_type, palette="dark")
fig = plt.gcf()
plt.show()
```

### *INSIGHT2-->We can see that the Entire Home/Apartment has the highest share, followed by the Private Room, and the least preferred is Shared Room.

```python
plt.figure(figsize=(13,10))
sns.heatmap(df.corr(method='kendall'), cmap='viridis', annot=True,vmin=-1,vmax=1)
plt.title('df corr()')

#can also be written like this
# corr = df.corr(method='kendall')
# plt.figure(figsize=(13,10))
# plt.title("Correlation Between Different Variables\n")
# sns.heatmap(corr, annot=True)
# plt.show()
```

### *INSIGHT3-->From the graph above, we know that there is not a strong correlation except review_per_month and number_of_review.

##  Neighbourhood Group vs. Availability Room

```python
plt.figure(figsize=(13,7))
plt.title("Room Type on Neighbourhood Group")
sns.countplot(df.neighbourhood_group,hue=df.room_type, palette="dark")
plt.show()
```

### *INSIGHT4-->The graph shows that the Entire Home/Apartment is listed most near Manhattan, while Private Rooms and Apartments Near Brooklyn are Nearly equal

##  Neighbourhood Group vs. Availability Room

```python
plt.style.use('classic')
plt.figure(figsize=(13,7))
plt.title("Neighbourhood Group vs. Availability Room")
sns.boxplot(data=df, x='neighbourhood_group',y='availability_365',palette="dark")
plt.show()
```

### *INSIGHT5-->The graph above shows the relationship between the availability room and neighborhood group.

## Neighbourhood Group Price Distribution

```python
plt.style.use('classic')
plt.figure(figsize=(13,7))
plt.title("Neighbourhood Group Price Distribution < 500")
sns.boxplot(y="price",x ='neighbourhood_group' ,data = df[df.price<500])
plt.show()
```

### *INSIGHT6-->From the boxplot above, we can definitely observe a couple of things about the distribution of prices for Airbnb in NYC.

We can state that Manhattan has the highest range price for the listings with about $140 as an average price, followed by Brooklyn with $90 per night,
Queens and Staten Island seem to have a very similar distribution,
The Bronx is the cheapest.
